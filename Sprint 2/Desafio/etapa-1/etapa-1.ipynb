```python
# Importação da biblioteca Pandas
import pandas as pd
```

```python
# Importação da biblioteca Matplotlib
import matplotlib.pyplot as plt
```

### Teste da biblioteca Pandas:

```python
dados = {
    "nome": ["Paulo", "Maria"],
    "idade": [24, 27]
}

df = pd.DataFrame(dados)

print("Biblioteca Pandas Instalada e Funcionando!")
display(df)
```

### Teste da biblioteca Matplotlib:

```python
nomes  =  ["Paulo", "Maria"]
idades = [24, 27]

plt.bar(nomes, idades, color="green")
plt.title("Teste da Biblioteca Matplotlib")
plt.xlabel("Nomes")
plt.ylabel("Idades")
plt.show()
```
